```python
x = np.arange(100)

fig, axs = plt.subplots(2,2)

axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine Wave")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")


axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Random Function")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Random Function")
axs[1, 1].set_xlabel("Test")

plt.tight_layout() ########### help to remove overlap maps
fig.suptitle("Four Plots")
plt.show()
```

```python
# 3D
ax = plt.axes(projection="3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
# line plot

ax = plt.axes(projection="3d")

x = np.arange(0, 50, 0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x, y, z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
#suface plot

ax = plt.axes(projection="3d")

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X , Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```

```python
dff = pd.read_csv(r'C:\Users\Rishi\Downloads\carr.csv')
dff.head()
```

1. Load cars data as dataframe using pandas and create a bar plot between
number of cylinders and frequency of cars with that many number of
cylinders

 Set xlabel as Number of cylinders
 Set ylabel as Frequency of cars
 Draw a bar plot

```python
plt.figure(figsize=(10, 6))
dff['cyl'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.xlabel('Number of Cylinders')
plt.ylabel('Frequency of cars')

plt.show()
```

2. Write code to load data from cars and print a bar graph of count of columns with null values.

```python
plt.figure(figsize=(12, 6))
dff.isnull().sum().plot(kind='bar', color='orange')
plt.xlabel('Columns')
plt.ylabel('Count of null values')

plt.show()
```

3. Use the 'mpg' (Miles Per Gallon column) and draw a histogram:
    
 Set xlabel: Miles per gallon
    
 Set ylabel: Frequency
    
 Set title as Miles Per Gallon Histogram

 Use mpg column to generate a histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(dff['mpg'], bins=20, color='green', edgecolor='black')

plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')

plt.show()
```

4. Draw a boxplot on the card dataframes hp column:
    
 Set xlabel: Car Horsepower
    
 Set title as Boxplot for car horsepower

 Use hp column to generate a boxplot


```python
plt.figure(figsize=(10,6))
plt.boxplot(dff['hp'], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))

plt.xlabel('Car Horsepower')
plt.title('Boxplot for Ca Horsepower')

plt.show()
```
